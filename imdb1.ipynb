```python
import requests,re,numpy as np
from bs4 import BeautifulSoup
```

```python
years=[]
titles=[]
run_time=[]
certificate=[]
rating=[]
genre=[]
votes=[]
meta_score = []
gross=[]
director=[]
for i in range(1,11):
#     https://www.imdb.com/list/ls006266261/?st_dt=&mode=detail&page=2&ref_=ttls_vm_dtl&sort=release_date,desc
    url = f'https://www.imdb.com/list/ls006266261/?st_dt=&mode=detail&page={i}&ref_=ttls_vm_dtl&sort=release_date,desc'
    request_header = {'Content-Type': 'text/html; charset=UTF-8','User-Agent': 'Chrome/101.0.0.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/119.0','Accept-Encoding': 'gzip, deflate, br'}
    page = requests.get(url,headers =request_header)
    page.text
    soup = BeautifulSoup(page.text)
#     for j in soup.find_all("div",class_='lister-item-content'):
#         m = j.find("span",class_="metascore favorable")
#         print(m)
#         ppp=m.text
# #         print(m)
#         mp = re.findall("(\d+)",ppp)
# #         print(mp)
#         if mp:
# #             print(mp)
#             meta_score.append(mp[0])
#         else:
#             meta_score.append(np.nan)  
            
    for j in soup.find_all("div",class_='lister-item-content'):
    #     print(i)
    #     print('*'*50)
        all1 = j.find("h3",class_='lister-item-header')
        all2 = all1.text.strip()
    #     print(all2)
        r1 = re.findall('[A-Za-z].+',all2)
        if r1:
            titles.append(r1[0])
        else:
            titles.append(np.nan)

        r2 = re.findall('\((\d+)\)',all2)
        if r2:
            years.append(r2[0])
        else:
            years.append(np.nan)

        c = j.find("span",class_='certificate')
        if c==None:
            certificate.append(np.nan)
        else:
            certificate.append(c.text)

        run_t = j.find("span",class_='runtime')
        r = run_t.text
        r1 = re.findall('(\d+)',r)
#         print(r1)
        if r1:
            run_time.append(r1[0])
        else:
            run_time.append(np.nan)

        ge= j.find("span",class_='genre')
        g = ge.text.strip()
        if g:
            genre.append(g)
        else:
            genre.append(np.nan)

        ra = j.find("span",class_='ipl-rating-star__rating')
        rat = ra.text.strip()
        if rat:
            rating.append(rat)
        else:
            rating.append(np.nan)
        
        
        dir11 = re.findall("Director:\n(\w.+)",j.text)
        if dir11:
            director.append(dir11[0])
        else:
            director.append(np.nan)
        

        grs = re.findall("Gross:\n\$(\d.+)M",j.text)
        if grs:
            gross.append(grs[0])
        else:
            gross.append(np.nan)

            
            
        g = j.find("span",{'name':'nv'})
        if g:
            votes.append(g.text)
        else:
            votes.append(np.nan)    
    
        m = j.find("span", class_="metascore favorable")

        # Check if the element is found
        if m:
            # Extract the text from the found element and remove leading/trailing whitespace
            metascore_text = m.text.strip()

            # Use regular expressions to extract the Metascore value
            mp = re.findall(r"(\d+)", metascore_text)

            if mp:
                # Append the Metascore value to the meta_score list
                meta_score.append(mp[0])
            else:
                # If no Metascore value is found, append NaN
                meta_score.append(np.nan)
        else:
            # If the span element is not found, append NaN
            meta_score.append(np.nan)
#     print(i)
```

```python
print(len(titles))
print(len(years))
print(len(run_time))
print(len(certificate))
print(len(genre))
print(len(meta_score))
print(len(rating))
print(len(votes))
print(len(director))
print(len(gross))
```
